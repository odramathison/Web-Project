```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

```python
def precio():
    driver = webdriver.Chrome(r"C:\Program Files\dChrome\chromedriver.exe")
    driver.get('https://es.investing.com/equities/')
    x=driver.find_element_by_id('stocksFilter')
    drop=Select(x)
    drop.select_by_visible_text('S&P 500')
    time.sleep(30)
    filas=driver.find_elements_by_tag_name('td')
    head=driver.find_elements_by_tag_name('th')

    data=[]
    for i in filas:
        data.append(i.text)

    columns_labels=[]
    for i in head:
        columns_labels.append(i.text)
    columns_labels=columns_labels[:10]

    driver.quit()
    n=10
    Datos=[data[i:i+n] for i in range(0,len(data),n)]
    global df
    df=pd.DataFrame(Datos[:490], columns=columns_labels)
    df=df.drop([' ', ''],axis=1)
    df.to_csv('acciones.csv', index=1, sep=',')
    return df
```
